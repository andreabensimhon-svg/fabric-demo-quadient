# Notebook 01 - Decouverte du Workspace Fabric

## Objectif
Comprendre ce qu est un **Workspace Fabric**, a quoi il sert, et decouvrir les differents types d items.

## Pour les anciens utilisateurs Cloudera
Un **Workspace Fabric** = votre ancien **cluster Cloudera** + **HDFS** + **Hive Metastore** + **Oozie**, le tout regroupe dans un seul environnement manage dans le cloud.

---

## 1. Qu est-ce qu un Workspace ?

Un **workspace** (espace de travail) est un **conteneur** dans Microsoft Fabric.

Imaginez un **dossier partage** dans lequel toute votre equipe peut :
- Stocker des donnees (Lakehouse)
- Ecrire du code (Notebooks)
- Creer des flux de traitement (Pipelines)
- Faire des rapports (Power BI)

### Correspondance Cloudera - Fabric

| Fabric | Equivalent Cloudera |
|--------|--------------------|
| **Lakehouse** | HDFS + Hive |
| **Warehouse** | Impala |
| **Notebook** | Zeppelin / Jupyter |
| **Data Pipeline** | Oozie / Airflow |
| **SQL analytics endpoint** | Hive JDBC |

## 2. Les roles d acces au workspace

| Role | Peut voir | Peut modifier | Peut partager | Peut supprimer |
|------|-----------|---------------|---------------|----------------|
| **Viewer** | Oui | Non | Non | Non |
| **Contributor** | Oui | Oui | Non | Non |
| **Member** | Oui | Oui | Oui | Non |
| **Admin** | Oui | Oui | Oui | Oui |

## 3. La capacite Fabric

La **capacite** c est le **moteur** qui fait tourner vos traitements.

- **Cloudera** : vous aviez X machines avec Y CPU et Z RAM
- **Fabric** : vous avez une capacite (F2, F4, F64) qui determine la puissance

| Capacite | CU | Usage typique |
|----------|----|--------------|
| F2 | 2 | Dev / Test |
| F8 | 8 | Petite equipe |
| F64 | 64 | Production |
