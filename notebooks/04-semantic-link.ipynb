## What is a Semantic Model?

A **semantic model** (formerly "dataset" in Power BI) is a business logic layer on top of your Lakehouse data:
- **Tables** and **relationships** (star/snowflake schema)
- **Measures** (DAX expressions: `Total Revenue := SUM(Sales[Amount])`)
- **Hierarchies** (Year → Quarter → Month)
- **RLS** (Row-Level Security policies)

### Why it matters
Without a semantic model, every consumer writes their own SQL/Spark queries and re-implements business logic. With a semantic model, calculations are **defined once** and reused across reports, notebooks, and APIs.

### Default Semantic Model
When you create a Lakehouse in Fabric, a **default semantic model** is auto-generated with the same name as the Lakehouse. It exposes all managed Delta tables.

> **For this demo**: Workspace = `demo-quadient`, Lakehouse = `demolakehouse`, Semantic Model = `demolakehouse`

## 1. Semantic Link — Bridging Data Engineering and BI

### The Problem
Traditionally, there's a wall between:
- **Data Engineers** working in notebooks (Python, Spark, SQL)
- **BI Analysts** working in Power BI (semantic models, DAX, reports)

Data engineers can't easily consume the measures and KPIs defined in Power BI.

### The Solution: Semantic Link
**Semantic Link** (`sempy`) is a Python library pre-installed in the Fabric Spark runtime that enables:
- **`list_datasets()`** — Enumerate all semantic models in the workspace
- **`list_tables(dataset)`** — Inspect table schema within a model
- **`read_table(dataset, table)`** — Load a table into a Pandas DataFrame
- **`evaluate_dax(dataset, dax_string)`** — Execute DAX queries server-side and return results as a DataFrame
- **`list_measures(dataset)`** — Retrieve all DAX measures defined in the model

### Key Benefit
Data scientists can reuse BI-curated, RLS-filtered data in ML pipelines without duplicating ETL logic or exporting CSVs.

````python
# ============================================
# STEP 1: Import Semantic Link
# ============================================
# sempy is pre-installed in Fabric Spark runtime — no pip install needed.

import sempy
import sempy.fabric as fabric

print(f"Semantic Link imported successfully.")
print(f"  sempy version: {sempy.__version__}")
print()
print("'sempy.fabric' is the main module for interacting with")
print("Power BI semantic models from a Spark notebook.")
````

````python
# ============================================
# STEP 2: List Semantic Models in the Workspace
# ============================================
# This lists all semantic models (datasets) the current user has access to
# in the current workspace.

import sempy.fabric as fabric

datasets = fabric.list_datasets()

print("Semantic models available in this workspace:\n")
if len(datasets) > 0:
    for idx, row in datasets.iterrows():
        print(f"  - {row['Dataset Name']}")
        print(f"    ID: {row['Dataset Id']}")
        print()
else:
    print("  No semantic models found.")
    print()
    print("  Troubleshooting:")
    print("  1. Open your Lakehouse 'demolakehouse' in Fabric")
    print("  2. Click on 'SQL analytics endpoint' in the top-right dropdown")
    print("  3. Go back to workspace view — the default semantic model should now appear")
    print("  4. If still missing: Lakehouse > 'New semantic model' > select tables > Save")
````

````python
# ============================================
# STEP 3: List Tables in the Semantic Model
# ============================================
# The default semantic model has the same name as the Lakehouse: "demolakehouse"
# It auto-exposes all managed Delta tables from the Lakehouse.

import sempy.fabric as fabric

DATASET_NAME = "demolakehouse"  # Default semantic model = Lakehouse name

try:
    tables = fabric.list_tables(DATASET_NAME)
    print(f"Tables in semantic model '{DATASET_NAME}':\n")
    for idx, row in tables.iterrows():
        print(f"  - {row['Name']}")
except Exception as e:
    print(f"Error: {e}")
    print()
    print(f"Could not find semantic model '{DATASET_NAME}'.")
    print("Make sure the default semantic model exists in your workspace.")
    print("Check Step 2 output for the exact model name available.")
````

````python
# ============================================
# STEP 4: Load a Table into a Pandas DataFrame
# ============================================
# read_table() pulls data from the semantic model (not directly from Delta).
# This means RLS policies and model-level security are respected.

import sempy.fabric as fabric

DATASET_NAME = "demolakehouse"
TABLE_NAME = "ventes_demo"  # Table created in Notebook 02

try:
    df = fabric.read_table(dataset=DATASET_NAME, table=TABLE_NAME)
    print(f"Loaded {len(df)} rows, {len(df.columns)} columns from '{TABLE_NAME}'")
    print(f"Columns: {list(df.columns)}")
    print()
    print(df.head(10))
    print()
    print("Note: This data comes from the Power BI semantic model,")
    print("not directly from the Delta table. RLS filters (if any) are applied.")
except Exception as e:
    print(f"Error: {e}")
    print()
    print(f"Could not read table '{TABLE_NAME}' from model '{DATASET_NAME}'.")
    print("Verify that:")
    print(f"  1. The semantic model '{DATASET_NAME}' exists (check Step 2)")
    print(f"  2. The table '{TABLE_NAME}' is exposed in the model (check Step 3)")
    print("  3. If using a custom semantic model, update DATASET_NAME above")
````

````python
# ============================================
# STEP 5: Execute a DAX Query
# ============================================
# DAX (Data Analysis Expressions) is the query language for Power BI models.
# evaluate_dax() runs the query server-side and returns results as a DataFrame.

import sempy.fabric as fabric

DATASET_NAME = "demolakehouse"

try:
    # DAX query: aggregate revenue by Filiale and Annee
    result = fabric.evaluate_dax(
        dataset=DATASET_NAME,
        dax_string="""
            EVALUATE
            SUMMARIZECOLUMNS(
                ventes_demo[Filiale],
                ventes_demo[Annee],
                \"Total_CA\", SUM(ventes_demo[CA]),
                \"Row_Count\", COUNTROWS(ventes_demo)
            )
        """
    )
    print("DAX query result (revenue by subsidiary and year):")
    print(result)
    print()
    print("The DAX query was evaluated server-side on the semantic model.")
    print("All model-level measures and security policies are respected.")
except Exception as e:
    print(f"Error: {e}")
    print()
    print("If this fails, verify:")
    print(f"  1. Model '{DATASET_NAME}' exists and has data")
    print("  2. Table 'ventes_demo' is exposed in the model")
    print("  3. Column names match exactly (Filiale, Annee, CA)")
````

````python
# ============================================
# STEP 6: Use Semantic Model Data for Analysis / ML
# ============================================
# Real-world use case: feed ML pipelines with BI-curated data
# without re-implementing business logic.

import sempy.fabric as fabric
import pandas as pd

DATASET_NAME = "demolakehouse"
TABLE_NAME = "ventes_demo"

try:
    df = fabric.read_table(dataset=DATASET_NAME, table=TABLE_NAME)

    print("Descriptive statistics on revenue (CA):")
    print("=" * 50)
    if 'CA' in df.columns:
        print(f"  Mean     : {df['CA'].mean():,.0f}")
        print(f"  Median   : {df['CA'].median():,.0f}")
        print(f"  Min      : {df['CA'].min():,.0f}")
        print(f"  Max      : {df['CA'].max():,.0f}")
        print(f"  Std Dev  : {df['CA'].std():,.0f}")
        print(f"  Count    : {len(df)}")
    else:
        print(f"  Column 'CA' not found. Available columns: {list(df.columns)}")

    print()
    print("This data came from the semantic model, not raw Delta.")
    print("Data scientists can now reuse BI-curated, RLS-filtered data")
    print("for ML workflows (scikit-learn, TensorFlow, etc.) without")
    print("duplicating ETL pipelines or exporting CSVs.")

except Exception as e:
    print(f"Error: {e}")
    print("Refer to troubleshooting steps in previous cells.")
````

## 2. Semantic Link — Function Reference

| Function | Purpose | Returns |
|----------|---------|--------|
| `fabric.list_datasets()` | List all semantic models in workspace | DataFrame (name, id) |
| `fabric.list_tables(dataset)` | List tables in a model | DataFrame (table names) |
| `fabric.read_table(dataset, table)` | Load table data | Pandas DataFrame |
| `fabric.evaluate_dax(dataset, dax)` | Execute DAX query | Pandas DataFrame |
| `fabric.evaluate_measure(dataset, measure, groupby)` | Evaluate a named measure | Pandas DataFrame |
| `fabric.list_measures(dataset)` | List all DAX measures | DataFrame (name, expression) |
| `fabric.list_relationships(dataset)` | List table relationships | DataFrame (FK definitions) |

## 3. Practical Implications for Quadient

### Use Case: BI ↔ Data Science Collaboration

```
BI Team creates a semantic model:
  - Tables: orders by subsidiary
  - Measures: revenue, margin, avg delivery time
  - RLS: each subsidiary sees only its own data

Data Science Team needs to:
  - Forecast parcel volumes for next month
  - Use the same data and calculations as BI

With Semantic Link:
  → 3 lines of Python to load BI-curated data
  → DAX measures reused server-side (no recalculation)
  → RLS enforced (no unauthorized data access)
```

### Without Semantic Link (legacy workflow)
1. Export data from Power BI to CSV
2. Reload into a notebook
3. Re-implement business logic
4. Hope the numbers match

### With Semantic Link
```python
df = fabric.read_table("demolakehouse", "ventes_demo")
# Done. BI-curated, RLS-filtered, ready for ML.
```

## 5. Troubleshooting

| Issue | Solution |
|-------|----------|
| `Dataset 'X' not found` | Verify the exact model name in `fabric.list_datasets()` output. The default model name matches the Lakehouse name (`demolakehouse`). |
| No semantic models listed | Open the Lakehouse → switch to SQL analytics endpoint view → return to workspace. The default model should appear. |
| Table not found in model | The table must exist in the Lakehouse **and** be exposed in the semantic model. For custom models, manually add the table. |
| DAX query error | Check column names match exactly (case-sensitive). Use `fabric.list_tables()` to verify. |
