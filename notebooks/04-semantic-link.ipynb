# Notebook 04 - Semantic Link (Bonus)

## Objectif
Decouvrir **Semantic Link**, une librairie Python pour interroger les **modeles semantiques Power BI** depuis un notebook Spark.

## C est quoi un modele semantique ?
C est une **couche metier** au-dessus de vos donnees : tables, relations, mesures (CA, marge...), hierarchies, securite.

---

## 1. Semantic Link - Le pont entre Data Engineering et BI

**Semantic Link** permet de :
- Lire les donnees depuis un modele semantique Power BI dans un DataFrame
- Executer des requetes DAX depuis un notebook
- Lister les modeles disponibles
- Alimenter des modeles ML avec des donnees nettoyees par la BI

### Fonctions principales

| Fonction | Ce qu elle fait |
|----------|----------------|
| fabric.list_datasets() | Liste les modeles semantiques |
| fabric.list_tables(dataset) | Liste les tables d un modele |
| fabric.read_table(dataset, table) | Charge une table en DataFrame |
| fabric.evaluate_dax(dataset, dax) | Execute une requete DAX |

```python
# ============================================
# DEMO : Importer Semantic Link
# ============================================
# Pre-installe dans Fabric, pas besoin de pip install

import sempy
import sempy.fabric as fabric

print(f"Semantic Link importe avec succes !")
print(f"Version : {sempy.__version__}")
```

```python
# ============================================
# DEMO : Lister les modeles semantiques
# ============================================

import sempy.fabric as fabric

datasets = fabric.list_datasets()

print("Modeles semantiques disponibles :")
if len(datasets) > 0:
    for idx, row in datasets.iterrows():
        print(f"  {row['Dataset Name']} - ID: {row['Dataset Id']}")
else:
    print("  Aucun modele trouve. Verifiez que votre Lakehouse a des tables.")
```

```python
# ============================================
# DEMO : Charger une table dans un DataFrame
# ============================================
# Remplacez DATASET_NAME par le nom de votre modele

import sempy.fabric as fabric

DATASET_NAME = "demo"  # A adapter

df = fabric.read_table(
    dataset=DATASET_NAME,
    table="ventes_demo"  # A adapter
)

print(f"Donnees chargees : {len(df)} lignes, {len(df.columns)} colonnes")
print(f"Colonnes : {list(df.columns)}")
print()
print(df.head(10))
```

```python
# ============================================
# DEMO : Executer une requete DAX
# ============================================

import sempy.fabric as fabric

DATASET_NAME = "demo"  # A adapter

result = fabric.evaluate_dax(
    dataset=DATASET_NAME,
    dax_string="""
        EVALUATE
        SUMMARIZECOLUMNS(
            ventes_demo[Filiale],
            \"Total_CA\", SUM(ventes_demo[CA]),
            \"Nb_Lignes\", COUNTROWS(ventes_demo)
        )
    """
)

print("Resultat de la requete DAX :")
print(result)
```

```python
# ============================================
# DEMO : Statistiques pour le Machine Learning
# ============================================

import sempy.fabric as fabric

DATASET_NAME = "demo"  # A adapter
df = fabric.read_table(dataset=DATASET_NAME, table="ventes_demo")

print("Statistiques descriptives :")
print("=" * 50)
if 'CA' in df.columns:
    print(f"  CA moyen    : {df['CA'].mean():,.0f}")
    print(f"  CA median   : {df['CA'].median():,.0f}")
    print(f"  CA min      : {df['CA'].min():,.0f}")
    print(f"  CA max      : {df['CA'].max():,.0f}")
print()
print("Les data scientists reutilisent les donnees nettoyees par la BI !")
```
