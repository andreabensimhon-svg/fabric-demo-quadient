# Notebook 02 - Lakehouse et OneLake

## Objectif
Comprendre le **Lakehouse** et **OneLake**, creer des tables, et comprendre comment les donnees sont organisees.

## Pour les anciens utilisateurs Cloudera
- **OneLake** = votre nouveau **HDFS**
- **Lakehouse** = **HDFS + Hive** fusionnes
- **Tables Delta** = vos **tables Hive** en mieux (ACID, versioning)

---

## 1. OneLake - Le Data Lake unifie

| HDFS (Cloudera) | OneLake (Fabric) |
|-----------------|------------------|
| Un cluster = un HDFS | Un tenant = un OneLake |
| Gere par vous (noeuds, disques) | Gere par Microsoft (serverless) |
| Replication manuelle | Replication automatique |

## 2. Structure du Lakehouse

```
Mon Lakehouse
  Tables/    <- Donnees structurees (format Delta)
  Files/     <- Fichiers bruts (CSV, JSON, Parquet)
```

````python
# ============================================
# DEMO : Creer une table de donnees de ventes
# ============================================
# Equivalent Cloudera : CREATE TABLE + INSERT INTO dans Hive

from pyspark.sql import Row

data = [
    Row(Id=1,  Annee=2024, Filiale="France",    Categorie="Courrier",   CA=150000),
    Row(Id=2,  Annee=2024, Filiale="France",    Categorie="Colis",      CA=230000),
    Row(Id=3,  Annee=2024, Filiale="Allemagne", Categorie="Courrier",   CA=180000),
    Row(Id=4,  Annee=2024, Filiale="Allemagne", Categorie="Colis",      CA=310000),
    Row(Id=5,  Annee=2024, Filiale="UK",        Categorie="Courrier",   CA=120000),
    Row(Id=6,  Annee=2024, Filiale="UK",        Categorie="Colis",      CA=195000),
    Row(Id=7,  Annee=2023, Filiale="France",    Categorie="Courrier",   CA=140000),
    Row(Id=8,  Annee=2023, Filiale="France",    Categorie="Colis",      CA=210000),
    Row(Id=9,  Annee=2023, Filiale="Allemagne", Categorie="Courrier",   CA=165000),
    Row(Id=10, Annee=2023, Filiale="Allemagne", Categorie="Colis",      CA=290000),
]

df = spark.createDataFrame(data)
df.write.mode("overwrite").saveAsTable("ventes_demo")

print("Table ventes_demo creee avec succes !")
print(f"Nombre de lignes : {df.count()}")
print(f"Colonnes : {df.columns}")
````

````python
# ============================================
# DEMO : Lire la table
# ============================================
# Exactement comme sur Cloudera avec Spark

df_ventes = spark.sql("SELECT * FROM ventes_demo")
df_ventes.show(truncate=False)

print("C est le meme code Spark que sur Cloudera !")
````

````python
# ============================================
# DEMO : Requetes analytiques avec Spark SQL
# ============================================

print("CA total par filiale et par annee :")
print("=" * 50)

spark.sql("""
    SELECT 
        Filiale, 
        Annee,
        SUM(CA) as CA_Total,
        COUNT(*) as Nb_Lignes
    FROM ventes_demo
    GROUP BY Filiale, Annee
    ORDER BY Filiale, Annee
""").show(truncate=False)

print("Meme syntaxe SQL que Hive. Aucun changement de code necessaire !")
````

````python
# ============================================
# DEMO : Charger un fichier CSV dans le Lakehouse
# ============================================
# Equivalent Cloudera : hdfs dfs -put fichier.csv /user/data/

from pyspark.sql import Row

files_path = "Files/demo"

df_csv = spark.createDataFrame([
    Row(Produit="Enveloppe C5", Prix=0.15, Stock=50000),
    Row(Produit="Enveloppe C4", Prix=0.25, Stock=30000),
    Row(Produit="Colis Standard", Prix=5.50, Stock=10000),
    Row(Produit="Colis Express", Prix=12.00, Stock=5000),
])

df_csv.write.mode("overwrite").option("header", True).csv(files_path)
print(f"Fichier CSV sauvegarde dans {files_path}")

df_from_csv = spark.read.option("header", True).csv(files_path)
df_from_csv.write.mode("overwrite").saveAsTable("produits_demo")
print("Table produits_demo creee a partir du CSV")
````

## Resume

| Concept | Ce qu il faut retenir |
|---------|----------------------|
| **OneLake** | Remplace HDFS. Un seul pour tout le tenant. |
| **Lakehouse** | Tables (Delta) + Files (bruts). |
| **Tables Delta** | Comme Hive, mais avec transactions ACID. |
| **SQL Endpoint** | Couche SQL instantanee en lecture. |
| **Code Spark** | Identique a Cloudera ! |
