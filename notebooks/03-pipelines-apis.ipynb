## 1. Data Pipelines

Un **pipeline** est une sequence d etapes automatisees :

```
Etape 1 : Recuperer les donnees brutes  -> Copy Data
Etape 2 : Nettoyer et transformer       -> Notebook
Etape 3 : Charger dans la table finale  -> Notebook
Etape 4 : Envoyer un email              -> Web
```

### Comparaison

| Oozie / Airflow | Data Pipeline Fabric |
|-----------------|---------------------|
| Fichier XML ou Python | Interface drag and drop |
| Serveur dedie a maintenir | Serverless |
| Actions : Spark, Hive, Shell | Activites : Notebook, Copy, SQL |

## 2. Les declencheurs (Triggers)

| Type | Description | Equivalent Cloudera |
|------|-------------|--------------------|
| **Manuel** | Clic ou appel API | oozie job -run |
| **Planifie** | Toutes les heures, tous les jours | Coordinator Oozie |
| **Evenement** | Quand un fichier arrive | Custom |

### Exemple pour Quadient
```
Tous les jours a 6h00 :
  1. Copier les nouvelles donnees
  2. Executer le notebook de transformation
  3. Rafraichir le rapport Power BI
```

## 3. Les APIs REST Fabric

Une **API REST** c est une **telecommande** pour controler Fabric depuis l exterieur.

| Action humaine | Commande API |
|----------------|-------------|
| Cliquer sur Executer | POST /pipelines/{id}/run |
| Regarder si le job est fini | GET /jobs/{id}/status |
| Lister les items | GET /workspaces/{id}/items |

````python
# ============================================
# DEMO : Obtenir un token d acces
# ============================================
# Dans un Notebook Fabric, on utilise mssparkutils

token = mssparkutils.credentials.getToken("https://api.fabric.microsoft.com")
print(f"Token obtenu ! (premiers caracteres : {token[:20]}...)")
print()
print("Ce token est comme votre badge d acces.")
````

````python
# ============================================
# DEMO : Lister les workspaces via l API
# ============================================

import requests

token = mssparkutils.credentials.getToken("https://api.fabric.microsoft.com")
url = "https://api.fabric.microsoft.com/v1/workspaces"
headers = {"Authorization": f"Bearer {token}"}

response = requests.get(url, headers=headers)

if response.status_code == 200:
    workspaces = response.json()["value"]
    print(f"{len(workspaces)} workspace(s) trouve(s) :")
    for ws in workspaces:
        print(f"  {ws['displayName']} - ID: {ws['id']}")
else:
    print(f"Erreur {response.status_code}: {response.text}")
````

````python
# ============================================
# DEMO : Lister les items d un workspace
# ============================================

import requests

token = mssparkutils.credentials.getToken("https://api.fabric.microsoft.com")
headers = {"Authorization": f"Bearer {token}"}

workspaces = requests.get(
    "https://api.fabric.microsoft.com/v1/workspaces",
    headers=headers
).json()["value"]

ws = workspaces[0]
ws_id = ws["id"]
print(f"Workspace : {ws['displayName']} (ID: {ws_id})")
print("-" * 50)

items = requests.get(
    f"https://api.fabric.microsoft.com/v1/workspaces/{ws_id}/items",
    headers=headers
).json()["value"]

print(f"{len(items)} item(s) :")
for item in items:
    print(f"  {item['displayName']} ({item['type']}) - ID: {item['id']}")
````

## 4. Tableau recapitulatif des APIs

| Action | Methode | URL |
|--------|---------|-----|
| Lister workspaces | GET | /v1/workspaces |
| Lister items | GET | /v1/workspaces/{id}/items |
| Declencher notebook | POST | /v1/workspaces/{id}/items/{id}/jobs/instances?jobType=RunNotebook |
| Declencher pipeline | POST | /v1/workspaces/{id}/items/{id}/jobs/instances?jobType=Pipeline |
| Suivre execution | GET | /v1/workspaces/{id}/items/{id}/jobs/instances/{runId} |

**Base URL** : https://api.fabric.microsoft.com

**Documentation** : https://learn.microsoft.com/en-us/rest/api/fabric/core/
